```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import csv

from sklearn import linear_model
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')

%config InlineBackend.figure_format = 'retina'
%matplotlib inline

plt.style.use('fivethirtyeight')
np.random.seed(42)
```

## Linear Regression Model

Linear Regression models allow us to study the relationships between two continuous variables (X, y). However, remember that the X variable is a combination of all our numerical and categorical (casted as numeric dummies) features. This will affect how this specific type of model performs.

### Step # 1: Load in the data 

Loading in the cleaned kaggle test data, the scaled training data and the standard scaler to use on our Linear Regression model.

```python
ames_df_test = pd.read_csv('../data/test.csv', index_col='Id')
X_train_sc = pd.read_csv('../data/X_train_sc.csv', header=None)
X_test_sc = pd.read_csv('../data/X_test_sc.csv', header=None)
y_train = pd.read_csv('../data/y_train.csv', header=None)
y_test = pd.read_csv('../data/y_test.csv', header=None)
ss = pd.read_pickle('../pickle/scaler.pkl')
```

### Quick check of the naive model

By taking the the mean of the y_train data we can get a sense for what the average of our predicted y variable should be close to at the end of our modeling process.

```python
np.mean(y_train)
```

### Step # 2: Instantiate the model

Identify the model we are going to be performing and assign it to a variable name for more efficient use.

```python
lr = LinearRegression()
```

### Step # 3: Fit the Model

Using the scaled training data from our previous notebook we'll fit this data to the type of model we've chosen to instantiate and use for predictions.

```python
lr.fit(X_train_sc, y_train)
```

### Step # 4: Score the Model

With the scaled training data fit to the model we've chose we can get scores that tell us about how well our model will perform when predicting.

- The score on our 75 % training data shows a pretty high score (very close to 1) which would tell us that we are looking to be a bit overfit
- The score on our 25 % testing data is an immediate red flag. The fact that we are showing a negative score tells us that we are trying to fit a model type that is not right for this data.

```python
lr.score(X_train_sc, y_train)
```

```python
lr.score(X_test_sc, y_test)
```

### Step # 5: Predict the Model

Using the trained scaled testing data predict the target variable (Sale Price).

```python
pred_test = lr.predict(X_test_sc)
```

```python
r2_score(y_test, pred_test)
```

```python
np.sqrt(mean_squared_error(y_test, pred_test)) ## RMSE
```

```python
plt.figure(figsize=(12,8))
sns.regplot(pred_test, y_test[0], color='red')
plt.title("GridSearch Test Predictions")
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
```

### Interpretation

The Linear Regression model is not the model to use on this data. If the graph above isn't enough to show how far off we are with the Linear Regression model then the Root Mean Squared Error of 952160673982196.2 should be the next metric to consult of how bad we are doing.
